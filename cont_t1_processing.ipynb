```python
import numpy as np 
import matplotlib.pyplot as plt
from numpy import random
import h5py
import os

from experiments.single_qubit import single_shot; 
```

```python
pth = "C:\\_Data\\LLqub_contT1\\"
fname= "t1_full2024-03-13_10-45-01.h5" 
g = h5py.File(pth+fname, 'r')
print(g.keys())
print(g['t1_recal'].keys())

t1_amp = g['t1_recal']['amp'][:]
t1_i = g['t1_recal']['i'][:]
t1_q = g['t1_recal']['q'][:]

xpts = g['t1_recal']['xpts'][:]

t1_amp_value = g['t1_recal']['amp_value'][:]
t1_i_value = g['t1_recal']['i_value'][:]
t1_q_value = g['t1_recal']['q_value'][:]
t1_amp_value_err = g['t1_recal']['amp_value_err'][:]
t1_i_value_err = g['t1_recal']['i_value_err'][:]
t1_q_value_err = g['t1_recal']['q_value_err'][:]

time_data = g['t1_recal']['time'][:]

```

Plot Raw T1 Continuous and Individual points 

```python
# create a subplot with 1 row and 2 columns
fig, ax = plt.subplots(1, 3, figsize=(10, 4))
ax[0].plot(xpts, t1_amp, '.')
ax[0].set_xlabel('Time [us]')
ax[0].set_ylabel('Amplitude [DAC Units]')
ax[0].set_title('Amp Value')

ax[1].plot(xpts, t1_i, '.')
ax[1].set_xlabel('Time [us]')
ax[1].set_ylabel('I [DAC Units]')
ax[1].set_title('I Value')

ax[2].plot(xpts, t1_q, '.')
ax[2].set_xlabel('Time [us]')
ax[2].set_ylabel('Q [DAC Units]')
ax[2].set_title('Q Value')

# format the subplots
plt.tight_layout()

plt.figure()

# add error bars to each of the points
fig, ax = plt.subplots(1, 3, figsize=(10, 4))

ax[0].errorbar(np.arange(len(t1_amp_value)), t1_amp_value, yerr=t1_amp_value_err, fmt = 'o', elinewidth= 0.5,  color = 'blue', label = 'Amp value')
ax[0].set_xlabel('Experiment Number')
ax[0].set_ylabel('T1 [us]') 
ax[0].set_title('Amp Value')

ax[1].errorbar(np.arange(len(t1_i_value)), t1_i_value, yerr=t1_i_value_err, fmt = 'o', elinewidth= 0.5,  color = 'blue', label = 'I value')
ax[1].set_xlabel('Experiment Number')
ax[1].set_ylabel('T1 [us]')
ax[1].set_title('I Value')

ax[2].errorbar(np.arange(len(t1_q_value)), t1_q_value, yerr=t1_q_value_err, fmt = 'o', elinewidth= 0.5,  color = 'blue', label = 'Q value')
ax[2].set_xlabel('Experiment Number')
ax[2].set_ylabel('T1 [us]')
ax[2].set_title('Q Value')

# format the subplots
plt.tight_layout()
```
